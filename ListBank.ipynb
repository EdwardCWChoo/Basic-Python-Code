```python
# --- View Contents function ---
def view_content ():
    # display content and content counts
    # if common list has no content, display message: No content found
    if common_list == []:
        print("\nNo content found.")
    # else display content separated by comma, and content counts
    else:
        list_str = common_list[0]
        for i in range(len(common_list)):
            if i == 0:
                pass
            else: 
                list_str += ", " + str(common_list[i])
        print("Contents:\n", list_str)
        print("Content counts: ", len(common_list))

# --- "Continue-or-Stop" prompt ---
def cont_prompt():
    while True:
        cont_input = input("Continue? [Y/N]: ")
        if cont_input.upper() == "Y" or cont_input.upper() == "N":
            break
        else: print("Invalid input.")
    return cont_input.upper()  # return value "Y" or "N"

# --- Add or Insert function ---
def add_insert():
    # prompt user to input either "A" or "I" only
    while True:
        subfunc_input = input('Enter "A" to add content or "I" to insert content: ')
        if subfunc_input.upper() == "A" or subfunc_input.upper() == "I":
            break
        else: print("Invalid input")
    
    # if enter "A", prompt user to enter content and append to the list
    if subfunc_input.upper() == "A":
        while True:
            new_content = input("Enter new content: ")
            common_list.append(new_content)
            cont_input = cont_prompt()    
            if cont_input == "Y":
                pass
            else: 
                break
    # else (or enter "I"), prompt user to enter content and position to be inserted
    else:
        while True:
            new_content = input("Enter new content: ")
            content_pos = input("Enter position to be inserted (0 as first item): ")
            common_list.insert(int(content_pos), new_content)
            cont_input = cont_prompt()
            if cont_input == "Y":
                pass
            else:
                break       

# --- Search and Modify / Delete function ---
def search_del_mod():
    search_func = ""
    if common_list == []:
        print("List is empty.")
    else:
        # prompt user to enter either "C" or "P"
        while search_func.upper() != "C" and search_func.upper() != "P":
            search_func = input('Enter "C" to search by content or "P" to search by position: ')
            if search_func.upper() != "C" and search_func.upper() != "P":
                print("Invalid input.")
            else: pass

        # if "C", prompt user for content input
        if search_func.upper() == "C":
            while True:
                content_input = input("Enter content: ")
                if content_input in common_list:
                    index = common_list.index(content_input)
                    found_content = common_list[index]
                    break
                else: print("Content not found")
        # else "P", prompt user for position input
        else:
            while True:
                pos_input = input("Enter position (0 as first): ")
                if pos_input.isdigit():
                    if int(pos_input) <= len(common_list)-1:
                        index = int(pos_input)
                        found_content = common_list[index]
                        break
                    else: print("invalid. Enter 0 to", len(common_list)-1)
                else: print("Content not found")

        print("Content found:", found_content)
        print("Content position:", index)

        # prompt user to enter either "D" or "M"
        while search_func.upper() != "D" and search_func.upper() != "M":
            search_func = input('Enter "D" to delete or "M" to modify: ')
            if search_func.upper() != "D" and search_func.upper() != "M":
                print("Invalid input.")
            else: pass

        # if "D", delete content from list
        if search_func.upper() == "D":
            common_list.pop(index)
            print('Succesfully deleted "%s" from the list.' % found_content)
        else:
            old_content = common_list[index]
            new_content = input("Enter new content to position #%d: " % index )
            common_list[index] = new_content
            print('Successfully replaced "%s" with "%s".' % (old_content, new_content))
        
# --- Export function ---
def export_file():
    file_name = input("Enter new file name: ")
    # arrange list into string array
    common_list_str = common_list[0]
    for i in range (len(common_list)):
        if i == 0:
            pass
        else:
            common_list_str += "\n" + str(common_list[i])
    
    new_file = open(file_name+".txt", "w+") # open file in write mode
    new_file.write(common_list_str)         # write content
    new_file.close()                        # close file
    
    print("Successfully exported to %s.txt" % file_name)

# --- Main Function ---
# define global function for list
common_list = []
data_choice = ""
print("Welcome to ListBank")
# prompt user if wanna start from new list or import list
while data_choice.upper() != "N" and data_choice.upper() != "I":
    data_choice = input('Enter "N" for start from new list or "I" for import list: ')
    if data_choice.upper() != "N" and data_choice.upper() != "I":
        print("Invalid input.")
    else: break

# if "N", do nothing
if data_choice.upper() == "N":
    print("Let's start with a new list!\n")
# if "I", prompt user for the file name, import file, clean file and assign to common_list
else:
    print("Please make sure that:\n- the file is .txt;\n- the file is located in the same folder;\n- the file name is exactly matched.")
    import_filename = input("Enter file name: ")
    import_file = open(import_filename+".txt", "r")
    import_list = import_file.readlines()
    import_file.close()
    
    for i in range (len(import_list)):
        temp = import_list[i].replace("\n","")
        import_list[i] = temp
    common_list = import_list
    print("Successfully imported list from %s.txt\n" % import_filename)
    
# prompt user to input correct function selection 
while True:
    # display function options and prompt for user's choice
    print("Select:\n1. View\n2. Add or Insert\n3. Search and Modify / Delete\n4. Export file\n5. Exit")
    while True:
        func_input = input("What would you like to do? ")
        if func_input.isdigit():
            if int(func_input) in range (1,6):
                break
            else: print("Invalid input.")
        else: 
            print("Invalid input.")

# call function accordingly
    if func_input == "1":
        view_content()
        print()
    elif func_input == "2":
        add_insert()
        print()
    elif func_input == "3":        
        search_del_mod()
        print()
    elif func_input == "4":
        export_file()
        print()
    else:
        print("\nThank you and see you again!")
        break
```
